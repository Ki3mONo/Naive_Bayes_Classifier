# Analiza Danych
W tym notebooku przeprowadzimy analizę danych dla zbiorów *Iris* oraz *Mushrooms*.  
W zbiorze *Iris* występują dane ilościowe, a w zbiorze *Mushrooms* dane kategoryczne.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Muchroom


```python
df_mushrooms = pd.read_csv("../data/mushrooms.csv")
missing_values = (df_mushrooms == '?').sum()
print("Brakujące wartości w każdej kolumnie:\n", missing_values)
df_mushrooms.head()
```

Kolumna danych stalk-root jest wybrakowana, występują brakujące wartości, więc na pewno pominiemy ją w analizie.

```python
class_counts = df_mushrooms['class'].value_counts()
print("Liczba grzybów w każdej klasie:\n", class_counts)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='class', hue='class', data=df_mushrooms, palette={'e': 'green', 'p': 'red'})
plt.title("Proporcja grzybów jadalnych i trujących")
plt.xlabel("Klasa (e: jadalny, p: trujący)")
plt.ylabel("Liczba przykładów")
plt.legend([], [], frameon=False)
plt.show()
```

```python
features = [col for col in df_mushrooms.columns if col != 'class']

for feature in features:
    crosstab = pd.crosstab(df_mushrooms[feature], df_mushrooms['class'])
    print(f"\nRozkład cechy {feature} względem klasy:\n", crosstab)

    plt.figure(figsize=(8, 6))
    crosstab.plot(kind='bar', stacked=True, color=['green', 'red'], ax=plt.gca())
    plt.title(f"Rozkład cechy {feature} w grzybach jadalnych i trujących")
    plt.xlabel(feature)
    plt.ylabel("Liczba grzybów")
    plt.legend(title="Klasa", labels=['jadalny (e)', 'trujący (p)'])
    plt.tight_layout()
    plt.show()
```

**Cechy, które uważamy za szczególnie ważne dla klasyfikacji:**
- bruises
- odor
- gill-spacing
- gill-size
- gill-color
- stalk-surface-above-ring
- stalk-surface-below-ring
- stalk-color-above-ring

- habitat

# Iris

```python
from sklearn.datasets import load_iris

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = [iris.target_names[i] for i in iris.target]

df_iris.head()
```

Widzimy, że w przypadku irysów mamy 4 cechy rozróżniające, wykorzystamy je wszystkie:
- sepal length (cm)
- sepal width (cm)
- petal length (cm)
- petal width (cm)

```python
features = iris.feature_names
sns.set(style="whitegrid")

for feature in features:
    plt.figure(figsize=(10, 4))
    sns.scatterplot(
        x=range(df_iris.shape[0]),
        y=df_iris[feature],
        hue=df_iris['species'],
        palette="deep",
        s=60
    )
    plt.title(f"Scatterplot dla cechy: {feature}")
    plt.xlabel("Indeks próbki")
    plt.ylabel(feature)
    plt.legend(title='Species')
    plt.show()
```
